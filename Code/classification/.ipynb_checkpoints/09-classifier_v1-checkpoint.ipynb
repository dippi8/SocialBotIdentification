```python
def oldness(x):
    if x[0] == '2':
        return 2018 - int(x[:4])
    else:
        return 2018 - int(x[-4:])
```

```python
def predict(X, y):
    forest = RandomForestClassifier(n_estimators = 5 , criterion = 'entropy', random_state = 42)
    logreg = linear_model.LogisticRegression()
    knn = KNeighborsClassifier(n_neighbors=10)
    scores = []
    scores.append(cross_val_score(forest, X, y, cv=10, scoring='f1_macro').mean())
    scores.append(cross_val_score(logreg, X, y, cv=10, scoring='f1_macro').mean())
    scores.append(cross_val_score(knn, X, y, cv=10, scoring='f1_macro').mean())
    return scores
```

```python
# only our descriptive features
X = full[['avg_fav', 'avg_hash', 'avg_len', 'avg_ret',
       'freq', 'max_fav', 'max_hash', 'max_len', 'max_ret', 'media_perc', 'min_fav', 'min_hash',
       'min_len', 'min_ret', 'quote_perc',
       'ret_perc', 'url_perc', 'description_len', 'name_len', 'screen_name_len', 'oldness']]
y = full['target']
```

```python
predict(X,y)
```

```python
forest.fit(X,y)
sorted(zip(map(lambda x: round(x, 4), forest.feature_importances_), X.columns), reverse=True)
```
