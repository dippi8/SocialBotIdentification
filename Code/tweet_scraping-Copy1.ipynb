## Using Selenium

```python
for userID in users:
    
    # initialize scraper
    browser = webdriver.Firefox()
    base_url = 'https://twitter.com/'
    try:
        # get user screen name
        user = api.get_user(userID)
        username = user.screen_name

        # scrape user tweets
        url = base_url + username
        browser.get(url)
        time.sleep(1)

        body = browser.find_element_by_tag_name('body')


        for _ in range(50):
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.2)

        tweets = browser.find_elements_by_class_name('tweet-text')

        for tweet in tweets:
            print(tweet.text)
    except:
        print("user not found")
    browser.close()
```

### Creazione df

```python
tweets = []
private_accounts = []
cont = 0

for user in users:
    try:
        stuff = api.user_timeline(screen_name = user, count = 100, include_rts = True)
        #tweet.is_quote_status indica se nel tweet c'è il riferimento ad un altro tweet

        for tweet in stuff:
            tweet._json['user_id'] = tweet._json['user']['id']
            if len(tweet._json['entities']['urls']) != 0:
                tweet._json['url'] = tweet._json['entities']['urls'][0]['expanded_url']
            else:
                 tweet._json['url'] = None
            del tweet._json['user'], tweet._json['entities']

            tweets.append(tweet._json)
    except:
        private_accounts.append(user)
        print(user + " is protected")
    cont += 1
    print(str(cont) + ' / 750' )
```

```python
tweets_df = pd.DataFrame.from_dict(tweets)
```

```python
# rimuove tutte le interruzioni (spero)

tweets_df['text'] = tweets_df['text'].apply(lambda x: x.replace('\r', ' '))
tweets_df['text'] = tweets_df['text'].apply(lambda x: x.replace('\n', ' '))
tweets_df['text'] = tweets_df['text'].apply(lambda x: x.replace('\t', ' '))
```
