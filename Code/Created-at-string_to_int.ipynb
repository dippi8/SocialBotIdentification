# Convert created_at string to day, month, year 

```python
import pandas as pd
import numpy as np
```

```python
tweets_df = pd.read_csv('data/tweets_df.csv', sep='\t')
```

```python
def get_month(month):
    if month == 'Jan':
        return 1
    elif month == 'Feb':
        return 2
    elif month == 'Mar':
        return 3
    elif month == 'Apr':
        return 4
    elif month == 'May':
        return 5
    elif month == 'Jun':
        return 6
    elif month == 'Jul':
        return 7
    elif month == 'Aug':
        return 8
    elif month == 'Sep':
        return 9
    elif month == 'Oct':
        return 10
    elif month == 'Nov':
        return 11
    elif month == 'Dec':
        return 12
    
```

```python
days, months, years = [], [], []

for row, tweet in tweets_df.iterrows():
    
    date = tweet['created_at'].split()
    days.append(int(date[2]))
    months.append(get_month(date[1]))
    years.append(int(date[5]))
    
    if row % 1000 == 0:
        print (str(row) + ' / 68068' )
```

```python
tweets_df['created_at_day'] = days
tweets_df['created_at_month'] = months
tweets_df['created_at_year'] = years
```

```python
tweets_df = tweets_df.drop('contributors', axis=1)
tweets_df = tweets_df.drop('coordinates', axis=1)
tweets_df = tweets_df.drop('extended_entities', axis=1)
tweets_df = tweets_df.drop('id_str', axis=1)
tweets_df = tweets_df.drop('created_at', axis=1)
```
